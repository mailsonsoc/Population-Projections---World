```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('UGEN-PPRJ_5001.csv')

Date = data['Date']
Genre = data['Male (Number)']

x=list(Date)
y=list(Genre)

plt.scatter(x,y)
plt.xlabel('Male')
plt.ylabel('Growth')
plt.title('Populations Projections - Male')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime

ugen = pd.read_csv('UGEN-PPRJ_5001.csv')

plt.figure(figsize=(8,5))
plt.title('Crescimento populacional')

x_coord = ugen['Date']
y1_coord = ugen['Female (Number)']
y2_coord = ugen['Male (Number)']
plt.xlabel('Years')
plt.ylabel('Growth')
plt.plot(x_coord, y1_coord, label = 'Female', marker = 'o')
plt.plot(x_coord, y2_coord, label = 'Male', marker = 'o')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt 
import pandas as pd 

data = pd.read_csv('UGEN-PPRJ_5001.csv')

df = pd.DataFrame(data)

X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])

plt.bar(X, Y, color='g')
plt.title("Populations Projections - Female")
plt.xlabel("Years")
plt.ylabel("Grownth")

plt.show()
```
